```python
exclude_phrases = ["Самый дешёвый", "Прилёт вечером", "Вылет утром",'Остался 1 билет по этой цене','Осталось 5 билетов по этой цене','Осталось 2 билета по этой цене','Осталось 3 билета по этой цене','Осталось 4 билета по этой цене']
# Пример перевода месяцев и дней недели
month_translation = {
    "янв": "Jan", "фев": "Feb", "мар": "Mar", "апр": "Apr",
    "май": "May", "июн": "Jun", "июл": "Jul", "авг": "Aug",
    "сен": "Sep", "окт": "Oct", "ноя": "Nov", "дек": "Dec"
}

day_translation = {
    "пн": "Mon", "вт": "Tue", "ср": "Wed",
    "чт": "Thu", "пт": "Fri", "сб": "Sat", "вс": "Sun"
}
```

```python
import csv
import re
from datetime import datetime, timedelta
import holidays
import os
import time
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.firefox import GeckoDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

options = webdriver.FirefoxOptions()
options.headless = True 

driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()), options=options)
```

```python
url='https://www.aviasales.kz/search/NQZ1712ALA1'
try:
    driver.get(url)
    wait = WebDriverWait(driver, 40)
    prices = wait.until(EC.presence_of_all_elements_located(
        (By.XPATH, "//div[@data-test-id='search-results-items-list']//div[contains(@class, 's__')]")
    ))
    print(len(prices))

except Exception as e:
    driver.save_screenshot('error_screenshot.png')
    print(f"Exception occurred: {e}")
```

```python


data = ''
prices = prices[1:]
data = '\n\n'.join(price.text for price in prices)
print(data)
```

```python
def translate_date(date_str):
    # Переводим месяцы
    for ru_month, en_month in month_translation.items():
        date_str = date_str.replace(ru_month, en_month)
    # Переводим дни недели
    for ru_day, en_day in day_translation.items():
        date_str = date_str.replace(ru_day, en_day)
    return date_str

def get_date_info(date_str):
    current_year = datetime.now().year  # Текущий год
    
    # Добавляем год в конец строки
    date_str_with_year = f"{date_str} {current_year}"
    
    # Переводим месяцы и дни недели на английский
    date_str_translated = translate_date(date_str_with_year)
    
    # Парсим дату с годом
    date_obj = datetime.strptime(date_str_translated, "%d %b, %a %Y")
    
    # Определяем праздники и выходные
    kz_holidays = holidays.Kazakhstan(years=current_year)
    is_weekend = date_obj.weekday() in [5, 6]  # Суббота = 5, воскресенье = 6
    is_holiday = date_obj in kz_holidays
    
    return date_obj.strftime("%Y-%m-%d"), is_holiday, is_weekend, date_obj


# Парсинг данных
tickets = []
blocks = data.strip().split("\n\n")  # Разделение по блокам

for block in blocks:

    lines = block.strip().split("\n")
    while lines and lines[0] in exclude_phrases:
        lines = lines[1:]

    # Исправленный парсинг цены с поддержкой всех пробелов
    price_match = re.search(r"\d{1,3}[\s\u202f ]\d{3}[\s\u202f ]₸", lines[0])
    if price_match:
        price = price_match.group(0).replace(" ", "").replace(" ", "").replace("₸", "")
    else:
        continue  # Пропускаем блок, если цена не найдена

    avialine = lines[4]
    from_time = lines[5]  # Время отправления
    from_city = lines[6]  # Астана
    date_info = lines[7]  # Дата отправления
    to_time = lines[11]   # Время прибытия
    to_city = lines[12]   # Алматы
    cleaned_duration = lines[9].replace("В пути:", "").replace(" ⁠", "").replace(" ", "").strip()
    match = re.search(r"(\d+)\s*ч\s*(\d+)\s*м", cleaned_duration)
    if match:
        hours = int(match.group(1))  # Часы
        minutes = int(match.group(2))  # Минуты
        duration = hours * 60 + minutes
    else:
        print("Неверный формат строки")

    # Получаем данные о дате
    curr_date, is_holiday, is_weekend, departure_date = get_date_info(date_info)

    # Рассчитываем дату прилета
    departure_datetime = datetime.strptime(f"{curr_date} {from_time}", "%Y-%m-%d %H:%M")
    arrival_datetime = datetime.strptime(f"{curr_date} {to_time}", "%Y-%m-%d %H:%M")

    # Если время прибытия меньше времени отправления, добавляем один день
    if arrival_datetime < departure_datetime:
        arrival_datetime += timedelta(days=1)

    # Добавляем билет в список
    tickets.append({
        "Price": price,
        "From": from_city,
        "To": to_city,
        "curr_date": curr_date,
        "departure_date": departure_datetime.strftime("%Y-%m-%d %H:%M"),
        "arrival_date": arrival_datetime.strftime("%Y-%m-%d %H:%M"),
        "is_holiday": is_holiday,
        "is_weekends": is_weekend,
        "time_interval": f"{from_time}-{to_time}",
        "avialine": avialine,
        "duration": duration
    })

# Запись в CSV
csv_filename = "nqzala.csv"
with open(csv_filename, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=["Price", "From", "To", "curr_date", "departure_date", 
                                              "arrival_date", "is_holiday", "is_weekends", 
                                              "time_interval", "avialine", "duration"])
    writer.writeheader()
    writer.writerows(tickets)

print(f"Данные успешно сохранены в файл: {csv_filename}")
```
